```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
# Use the print() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(coordinate[0], coordinate[1])
```

```python
from citipy import citipy
```

```python
# Create a list for holding the cities:
cities=[]

# Identify the nearest city for each lat long combo:

for coordinate in coordinates:
    city = (citipy.nearest_city(coordinate[0], coordinate[1]).city_name)
    
    # If the city is unique, then we will add it to the cities list:
    
    if city not in cities:
        cities.append(city)
        
#Print city count
len(cities)
```

```python
#Import requests
import requests

#Import API file MAKE SURE THE FILE NAME config and the api key variable name are printed correctly)
from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# Import the time library and the datetime module from the datetime library:

import time
from datetime import datetime
```

```python
# Create an empty list to hold the weather data.
city_data = []

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

# Create counters.
record_count = 1
set_count = 1
```

```python
#Loop through all the cities: in the list:

for i, city in enumerate(cities):
    
      
    if (i % 50 == 0 and i>= 50):
        set_count+=1
        record_count=1
        time.sleep(60)
    
    #Create an endpoint URL with each city:
    
    city_url = url + "&q=" + city.replace(" ", "+")
    
    # Log your URL, record and set numbers and city
    
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    #Add 1 to record count
    
    record_count +=1
    
    
    # Run an API request for each of the cities.

    try:
        # Parse the JSON and retrieve data.    
        city_weather = requests.get(city_url).json()

        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date= city_weather["dt"]
        city_name= city_weather["name"]

            # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_date).strftime('%Y-%m-%d %H:%M:%S')
            # Append the city information into city_data list.
        city_data.append({"City": city_name,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    # If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

    # Indicate that Data Loading is complete.
    print("-----------------------------")
    print("Data Retrieval Complete      ")
    print("-----------------------------")
```

```python
#Convert array of dictionaries into dataframe:

city_data_df=pd.DataFrame(city_data)
city_data_df
```

```python
#Reorder the columns:
new_order=["City", "Country", "Date", "Lat", "Lng","Max Temp", "Humidity", "Cloudiness", "Wind Speed"]

#Assign the new columns to dataframe:
city_data_df=city_data_df[new_order]
city_data_df.head()
```

```python
#Create the output file:

output_data_file ='Weather Data/cities.csv'

#Export city data into csv

city_data_df.to_csv(output_data_file, index_label="City ID")
```

### Linear Regression

```python
#Create northern and southern hemispheres:
```

```python
north_hemi_df= city_data_df.loc[(city_data_df['Lat'] >=0)]
south_hemi_df= city_data_df.loc[(city_data_df['Lat'] <0)]
```

```python
from scipy.stats import linregress

# Create a function to create perform linear regression on the weather data
# and plot a regression line and the equation with the data.

def plot_linear_regression(x_values, y_values, title, y_label, text_coordinates):
   
    # Run regression on hemisphere weather data.
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)

    # Calculate the regression line "y values" from the slope and intercept.
    regress_values = x_values * slope + intercept
    
    # Get the equation of the line.
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    # Create a scatter plot and plot the regression line.
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r")
    
    # Annotate the text for the line equation.
    plt.annotate(line_eq, text_coordinates, fontsize=15, color="red")
    plt.title(title)
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    plt.show()
```

```python
# Linear Regression on Max Temp for Northern Hemisphere:

x_values=north_hemi_df["Lat"]
y_values=north_hemi_df["Max Temp"]
 

#Call the funtion:
plot_linear_regression(x_values, y_values, 
                       'Linear Regression on the Northern Hemisphere for Maximum Temperature', 'Max Temp', (4, 35))

```

```python
# Linear Regression on Max Temp for Southern Hemisphere:

x_values=south_hemi_df["Lat"]
y_values=south_hemi_df["Max Temp"]
 

#Call the funtion:
plot_linear_regression(x_values, y_values, 
                       'Linear Regression on the Southern Hemisphere for Maximum Temperature', 'Max Temp', (-55, 90))

```

#### FINDING:

Correlation between *latitude* and *max temp* is strong.  
From the data:  
Closer to the equator: Warmer temp  
Further from the equator: Colder temp

```python
#Linear Regression on % Humidity for Northern Hemisphere :

x_values=north_hemi_df["Lat"]
y_values=north_hemi_df["Humidity"]
 

#Call the funtion:
plot_linear_regression(x_values, y_values, 
                      'Linear Regression on the Northern Hemisphere for % Humidity',
                       '% Humidity', (45, 10))

plt.show()
```

```python
#Linear Regression on % Humidity for Southern Hemisphere :

x_values=south_hemi_df["Lat"]
y_values=south_hemi_df["Humidity"]
 

#Call the funtion:
plot_linear_regression(x_values, y_values, 
                      'Linear Regression on the Southern Hemisphere for % Humidity',
                       '% Humidity', (-55, 10))

plt.show()
```
